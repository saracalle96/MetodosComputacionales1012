Una persona conduce una moto con rapidez constante vm
, mientras que un policia lo hace con aceleración constante ap
 (iniciando desde el reposo). Suponga que ambos movimientos inician en t=0
, haga una gráfica para el movimiento de la moto y del policia, donde el usuario ingrese la velocidad (moto) y aceleración (policia). Además determine el instante t
 en el cual se encuentran ambos. Ambos movimientos iniciarán en xi=0

Pruebe: para vm
 = 15m/s y ap
 = 3 m/s², el tiempo deberia ser 10 s

```python
import numpy as np
import matplotlib.pyplot as plt
def calcular_encuentro(vm,ap):
  return 2*vm/ap
def graficar_movimiento(vm,ap):
  t_encuentro=calcular_encuentro(vm,ap)
  t=np.linspace(0, 1.5*t_encuentro, 1000)
  x_moto=vm*t
  x_policia=0.5*ap*t**2
  plt.figure(figsize=(10,6))
  plt.plot(t,x_moto,label=f'Moto(v={vm})',color='blue')
  plt.plot(t, x_policia, label=f'Policía (a={ap} m/s²)', color='red')
  plt.scatter(t_encuentro, vm * t_encuentro, color='green', zorder=5,
                label=f'Encuentro en t={t_encuentro:.2f} s')
  plt.xlabel('Tiempo (s)')
  plt.ylabel('Posición (m)')
  plt.title('Movimiento de la moto y del policía')
  plt.legend()
  plt.grid(True)
  plt.show()
  return t_encuentro
vm = 15  # m/s
ap = 3   # m/s²

print(f"Calculando tiempo de encuentro para vm={vm} m/s y ap={ap} m/s²...")
t = graficar_movimiento(vm, ap)
print(f"Tiempo de encuentro calculado: {t} segundos")

def graficar_movimiento(vm,ap):
  t_encuentro=calcular_encuentro(vm,ap)
  t=np.linspace(0, 1.5*t_encuentro, 1000)
  x_moto=vm*t
  x_policia=0.5*ap*t**2


  plt.plot(t, x_policia, label=f'Policía (a={ap} m/s²)', color='red')
  plt.scatter(t_encuentro, vm * t_encuentro, color='green', zorder=5,
                label=f'Encuentro en t={t_encuentro:.2f} s')
  plt.xlabel('Tiempo (s)')
  plt.ylabel('Posición (m)')
  plt.title('Movimiento de la moto y del policía')
  plt.legend()
  plt.grid(True)
  plt.show()
vm = 15  # m/s
ap = 3   # m/s²

print(f"Calculando tiempo de encuentro para vm={vm} m/s y ap={ap} m/s²...")
t = graficar_movimiento(vm, ap)
print(f"Tiempo de encuentro calculado: {t} segundos")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def lennard_jones_potencial(r, epsilon, sigma):
  #calcula el potencial de lennard-jones para una distancia r
  term1=(sigma/r)**12
  term2=(sigma/r)**6
  V=4*epsilon*(term1-term2)
  return V
#Ejemplo de uso
epsilon=1
sigma=1
r_values = np.linspace(0.9 * sigma, 3 * sigma, 500)
V_values = lennard_jones_potencial(r_values,epsilon=1.0,sigma=1.0)
#grafico
plt.figure(figsize=(8, 5))
plt.plot(r_values,V_values,label='potencial de Lennard-Jones',color='blue')
plt.xlabel('Distancia (r)')
plt.ylabel('Potencial (V(r))')
plt.title('Potencial de Lennard-Jones')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
def lennard_jones_potencial(r, epsilon, sigma,r_cutoff=2.5):
  """Potencial de lennnard jones"""
  V=4*epsilon*((sigma/r)**12-(sigma/r)**6)

  V[r > r_cutoff]=0
  return V
sigma=1
epsilon=1
r_cutoff=2.5*sigma
r_values = np.linspace(0.9 * sigma, 3 * sigma, 500)
V_values = lennard_jones_potencial(r_values,epsilon,sigma, r_cutoff)
#grafico
plt.figure(figsize=(10, 6))
plt.plot(r_values, V_values, label=f'Potencial truncado en $r_c={r_cutoff}\sigma$', color='blue')
plt.axvline(r_cutoff, color='red', linestyle='--', label=f'$r_c = {r_cutoff}\sigma$')
plt.axvline(2**(1/6), color='green', linestyle=':', label='$r_m = 2^{1/6}\sigma$')
plt.axhline(0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel('Distancia ($r/\sigma$)')
plt.ylabel('Potencial ($V(r)/\epsilon$)')
plt.title('Potencial de Lennard-Jones (truncado)')
plt.legend()
plt.grid(True)
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar y limpiar los datos
url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"
alturas = pd.read_html(url)[0]

# Eliminar columnas en pulgadas y renombrar
alturas = alturas.drop(columns=[col for col in alturas.columns if 'in' in col])
alturas.columns = ['Rango', 'País', 'Altura Hombres (cm)', 'Altura Mujeres (cm)']

# 2. Ordenar por altura de hombres (podría ser mujeres también)
alturas_sorted = alturas.sort_values(by='Altura Hombres (cm)', ascending=False)

# 3. Seleccionar los 10 más altos, los 10 más bajos y agregar Colombia
top_10 = alturas_sorted.head(10)
bottom_10 = alturas_sorted.tail(10)
colombia = alturas[alturas['País'] == 'Colombia']

# Combinar los datos para el gráfico
comparacion = pd.concat([top_10, bottom_10, colombia]).drop_duplicates()

# 4. Graficar
plt.figure(figsize=(14, 8))

# Barras para hombres (azul)
plt.barh(
    comparacion['País'],
    comparacion['Altura Hombres (cm)'],
    color='blue',
    alpha=0.7,
    label='Hombres'
)

# Barras para mujeres (rojo)
plt.barh(
    comparacion['País'],
    comparacion['Altura Mujeres (cm)'],
    color='red',
    alpha=0.7,
    label='Mujeres'
)

plt.xlabel('Altura (cm)')
plt.title('Comparación de altura media entre hombres y mujeres (19 años)')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
